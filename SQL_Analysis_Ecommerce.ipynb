#  SQL Analysis of E-commerce Business Performance 


This project analyzes an e-commerce dataset using SQL to evaluate revenue trends, product performance, and customer behavior.

## Business Questions

1. How is revenue trending over time?
2. Which product categories generate the most revenue?
3. Who are the top customers by number of orders?
4. What percentage of orders are returned or canceled?

## Query 1 - Revenue Over Time


```python
query = """
SELECT
  DATE(created_at) AS order_date,
  SUM(sale_price) AS total_revenue
FROM
  `bigquery-public-data.thelook_ecommerce.order_items`
GROUP BY
  order_date
ORDER BY
  order_date;
"""
```

**Insight:**
Revenue fluctuates over time, showing daily variation in customer purchasing behavior. These patterns can help identify peak sales periods
and support short-term forecasting decisions.

## Query 2 - Revenue by Product Category

```python
query = """
SELECT
  p.category AS product_category,
  SUM(oi.sale_price) AS total_revenue
FROM
  `bigquery-public-data.thelook_ecommerce.order_items` AS oi
JOIN
  `bigquery-public-data.thelook_ecommerce.products` AS p
ON
  oi.product_id = p.id
GROUP BY
  product_category
ORDER BY
  total_revenue DESC;
  """
```

**Insight:** The analysis shows that some product categories generate significantly more revenue than others. 
    This highlights the most important categories driving sales and helps the business focus on its highest-value products.

## Query 3 - Customer Behavior

```python
query = """
SELECT
  user_id,
  COUNT(order_id) AS total_orders
FROM
  `bigquery-public-data.thelook_ecommerce.orders`
GROUP BY
  user_id
ORDER BY
  total_orders DESC
LIMIT 10;
"""
```

**Insight:**
The analysis identifies the most active customers based on total order count. A small group of users accounts for a higher number of purchases, indicating potential loyal or repeat customers. These customers are valuable for retention strategies, targeted marketing, and loyalty programs.

## Query 4 - Operational Insight (order Status)

```python
query = """
SELECT
  status,
  COUNT(order_id) AS total_orders
FROM
  `bigquery-public-data.thelook_ecommerce.orders`
GROUP BY
  status
ORDER BY
  total_orders DESC;
  """
```

**Insight:** The analysis shows the distribution of order statuses across the platform. Completed orders make up the majority, indicating healthy fulfillment performance. However, the presence of cancelled or returned orders highlights potential issues in customer satisfaction, logistics, or product expectations. Monitoring these trends helps improve operational efficiency and customer experience.
